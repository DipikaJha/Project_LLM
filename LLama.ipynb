```python
import json

# Dummy Q&A pairs
data = [
    {"instruction": "What is the purpose of Decentraland?", "output": "Decentraland is a decentralized virtual world where users can buy virtual land, build experiences, and interact with others using blockchain-based ownership."},
    {"instruction": "How do I purchase land in Decentraland?", "output": "To purchase land in Decentraland, visit the official Decentraland marketplace, connect your Ethereum wallet, and place a bid or buy available parcels using MANA tokens."},
    {"instruction": "Explain NFTs in simple terms.", "output": "NFTs, or Non-Fungible Tokens, are digital assets that represent ownership of unique items like art, music, or virtual real estate, stored securely on a blockchain."},
    {"instruction": "Why are NFTs valuable?", "output": "NFTs are valuable because they provide proof of ownership, scarcity, and authenticity for digital assets, which can be important for collectors, investors, and creators."},
    {"instruction": "What is a blockchain?", "output": "A blockchain is a digital ledger of transactions that is decentralized, transparent, and secure. It stores data across a network of computers."},
    {"instruction": "What is Ethereum used for?", "output": "Ethereum is a blockchain platform used for creating decentralized applications and smart contracts. It also supports NFTs and DeFi platforms."},
    {"instruction": "How can I earn in Decentraland?", "output": "You can earn in Decentraland by selling virtual land, creating and selling wearables, hosting events, or offering services like games or experiences."},
    {"instruction": "Can I build in Decentraland without coding?", "output": "Yes, Decentraland offers a drag-and-drop Builder tool that allows you to create scenes and environments without writing code."},
    {"instruction": "What is MANA token?", "output": "MANA is Decentraland’s native cryptocurrency. It is used to buy land, wearables, and services within the platform."},
    {"instruction": "Is Decentraland free to use?", "output": "Yes, anyone can enter and explore Decentraland for free. However, buying land or digital goods requires MANA tokens."}
]

# Save to a JSONL file
with open("llama_dummy_dataset.jsonl", "w") as f:
    for item in data:
        json.dump(item, f)
        f.write("\n")

print("✅ JSONL file saved as 'llama_dummy_dataset.jsonl'")
```
